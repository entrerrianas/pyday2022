## Interactuamos en forma dinámica con algunos valores de nuestras funciones

````python
import ipywidgets as widgets
from ipywidgets import HBox, VBox
from IPython.display import display
````

````python
#dominio
x = np.linspace(-50, 50)


@widgets.interact(n=(0, 9))
def p(n):
    y =  x**n #codominio con n variando entre 0 y 9
    grafica.plot(x, y)
    grafica.grid(True)
    grafica.show()
````

 ### <div class="alert alert-warning">Desplazamientos verticales:</div>

````python
x = np.linspace(-50, 50)


@widgets.interact(n=(0, 1000,250))
def p(n):
    y = n+x**2
    grafica.plot(x, y)
    grafica.grid(True)
    grafica.show()
````

 ### <div class="alert alert-warning">Desplazamientos horizontales:</div>

````python
@widgets.interact(n=(-20, 20))
def p(n):
    x = np.linspace(n-50, n+50)
    y = (x-n)**2
    grafica.plot(x, y)
    grafica.grid(True)
    grafica.show()
````

````python
x_2 = np.linspace(-5, 5)
x_1_4 = np.linspace(-10, 10)
x_4 = np.linspace(-2.5, 2.5)
grafica.figure(figsize=(10,8))
grafica.plot(x_2, x_2**2, linewidth=3, label='x^3')
grafica.plot(x_1_4, 1/4*x_1_4**2, linestyle='dashed' , linewidth=3, label='1/4*x^2')
grafica.plot(x_4, 4*x_4**2, linestyle='dotted' , linewidth=3, label='4*x^2')
grafica.legend()
grafica.grid(True)
grafica.show()
````

````python
@widgets.interact(grado=(0, 9), N_puntos=(5,35))
def mi_plot(grado=3, N_puntos=5):    
    x = np.linspace(-10, 10, N_puntos)
    y = x**grado
    grafica.figure(figsize=(12,8))
    grafica.plot(x, y, 'ro-')
    grafica.grid(True)
    grafica.show()
````

## Generar - mostrar un gif

````python
import random
from matplotlib import pyplot as plt
import gif

x = [random.randint(0, 100) for _ in range(100)]
y = [random.randint(0, 100) for _ in range(100)]

@gif.frame
def plot(i):
    xi = x[i*10:(i+1)*10]
    yi = y[i*10:(i+1)*10]
    plt.scatter(xi, yi)
    plt.xlim((0, 100))
    plt.ylim((0, 100))

frames = []
for i in range(10):
    frame = plot(i)
    frames.append(frame)

gif.save(frames, 'example-gif-01.gif', duration=100)
````
